## Bivariate Exploration

> first i will look at relation between DepDelay and the airport that delay occured in it and that will make us to know the top airports thoes delay occuerd in it.

```python
# i found negtive delays and that means some flights departure from airport before the scheduled
df_joined_has_delays = df_joined["DepDelay"] = df_joined[df_joined["DepDelay"] > 0]
```

```python
# check
df_joined_has_delays
```

```python
# groupby airport and sum Delay in this airport throgh 2008 and get top 10 
df_joined_has_delays_grouped = df_joined_has_delays.groupby(['airport_origin'], as_index=False).agg({'DepDelay': "sum"}).nlargest(10,"DepDelay")
df_joined_has_delays_grouped
```

```python
# check
df_joined_has_delays_grouped.shape
```

```python
# drop the index
df_joined_has_delays_grouped.reset_index(drop=True, inplace=True)
```

```python
#check
df_joined_has_delays_grouped
```

```python
# plot the relation between Depdelay and airport origin
plt.figure(figsize=[15, 5])
sb.set_theme(style="whitegrid")
sb.barplot(y="airport_origin", x="DepDelay", data=df_joined_has_delays_grouped);
```
